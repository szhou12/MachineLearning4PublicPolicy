```python
def get_best_models(result_df, metric):
    if metric == 'train_time' or metric == 'test_time':
        temp = result_df.sort_values(by='train_time', ascending=True)
        best_by_metric = temp.groupby('model', as_index=False).first().sort_values(by='train_time', ascending=True)
    else:
        temp = result_df.sort_values(by='train_time', ascending=False)
        best_by_metric = temp.groupby('model', as_index=False).first().sort_values(by='train_time', ascending=False)
    return best_by_metric
```
