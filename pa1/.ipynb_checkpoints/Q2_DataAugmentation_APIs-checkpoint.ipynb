```python
import numpy as np
import pandas as pd
import requests
```

```python
build_raw = pd.read_csv("data/311_Service_Requests_-_Vacant_and_Abandoned_Buildings_Reported.csv")
light_raw = pd.read_csv("data/311_Service_Requests_-_Alley_Lights_Out.csv")
```

```python
build = build_raw.dropna(subset=['ADDRESS STREET NUMBER', 'ADDRESS STREET DIRECTION','ADDRESS STREET NAME', 'ADDRESS STREET SUFFIX'])
light = light_raw[light_raw['Status'].isin(['Completed', 'Open'])]
```

```python
def address_update(address_list):
    new_address = []
    for i in address_list:
        temp = i.split()
        i = '+'.join(temp)
        new_address.append(i)
    return new_address
```

```python
def request_tract(update_address):
    by_tract = []
    for street in update_address:
        url='https://geocoding.geo.census.gov/geocoder/geographies/address?street={}&city=Chicago&state=IL&benchmark=Public_AR_Census2010&vintage=Census2010_Census2010&layers=14&format=json'.format(street)
        r = requests.get(url)
        rv = r.json()
        try:
            county = rv['result']['addressMatches'][0]['geographies']['Census Blocks'][0]['COUNTY']
            tract = rv['result']['addressMatches'][0]['geographies']['Census Blocks'][0]['TRACT']
        except:
            county = None
            tract = None
        by_tract.append((county, tract))
    return by_tract
```

```python
def request_acs(by_tract):
    colnames = ('Total Population', 'White', 'Male', 'Median Family Income', 'Average Household Size','State','County','Tract')
    acs_df = pd.DataFrame(columns=colnames)
    for i in by_tract:
        county = i[0]
        tract = i[1]
        if county != None:
            url ='https://api.census.gov/data/2015/acs5?get=B01003_001E,B02001_002E,B05003_002E,B19113_001E,B25010_001E&for=tract:{}&in=state:17%20county:{}&key=9d0f16e1040555cc09541467125125184391eed1'.format(tract, county)
            r = requests.get(url)
            rv = r.json()
            values = [rv[1]]
        else:
            values = [[np.nan]*len(colnames)]
        df_temp = pd.DataFrame(values, columns=colnames)
        acs_df = acs_df.append(df_temp, ignore_index=True)
    return acs_df
    
    
```

```python
cutoff = -3000
```

```python
light_address = light['Street Address'].tolist()
light_address_upate = address_update(light_address)
light_by_tract = request_tract(light_address_upate[cutoff:])
light_acs_df = request_acs(light_by_tract)
```

```python
light_select = light[cutoff:].reset_index(drop=True)
light_aug = pd.concat([light_select, light_acs_df], axis=1)
light_aug = light_aug.dropna(subset=['Total Population', 'White', 'Male', \
                                     'Median Family Income', 'Average Household Size',\
                                     'State','County','Tract'])
```

```python
build_broken_address = build[['ADDRESS STREET NUMBER', 'ADDRESS STREET DIRECTION',\
                              'ADDRESS STREET NAME', 'ADDRESS STREET SUFFIX']]
seam = build_broken_address.astype(str).apply(lambda x: ' '.join(x), axis=1)
build_address = seam.tolist()
build_address_upate = address_update(build_address)
build_by_tract = request_tract(build_address_upate[cutoff:])
build_acs_df = request_acs(build_by_tract)
```

```python
build_select = build[cutoff:].reset_index(drop=True)
build_aug = pd.concat([build_select, build_acs_df], axis=1)
build_aug = build_aug.dropna(subset=['Total Population', 'White', 'Male', 'Median Family Income', \
                                     'Average Household Size','State','County','Tract'])
```

```python
def rank_by_tract(acs_df, data_aug):
    acs_df_modified = acs_df.dropna(subset=['Total Population', 'White', 'Male', 'Median Family Income', \
                                            'Average Household Size','State','County','Tract'])
    data_temp = data_aug['Tract'].value_counts().to_frame().reset_index()
    data_temp.columns = ['Tract', 'Counts']
    data_rank = pd.merge(data_temp, acs_df_modified, on='Tract')
    data_rank = data_rank.drop_duplicates(['Tract'])
    data_rank = data_rank.reset_index(drop=True)    
    return data_rank
```

```python
# Types of blocks get 'Vacant/Abandoned Buildings Reported'
build_rank = rank_by_tract(build_acs_df, build_aug)
build_rank[:10]
```

```python
# Types of blocks get 'Alley Light Out'
light_rank = rank_by_tract(light_acs_df, light_aug)
light_rank[:10]
```

```python
# Difference in blocks
br = build_rank
lr = light_rank
```

```python
build_white_pct = br['White'].astype(int).mean()/br['Total Population'].astype(int).mean()
build_male_pct = br['Male'].astype(int).mean()/br['Total Population'].astype(int).mean()
light_white_pct = lr['White'].astype(int).mean()/lr['Total Population'].astype(int).mean()
light_male_pct = lr['Male'].astype(int).mean()/lr['Total Population'].astype(int).mean()
```

```python
# male percentage and white percentage in blocks that get 'Vacant/Abandoned Buildings Reported' vs 'Alley Light Out'
mw_df = pd.DataFrame({'White': [build_white_pct, light_white_pct],\
                    'Male': [build_male_pct, light_male_pct]},\
                    index=['Vacant/Abandoned Buildings Reported', 'Alley Light Out'])
mw_df
```

```python
build_income = br['Median Family Income'].astype(float).mean()
build_house_size = br['Average Household Size'].astype(float).mean()
light_income = lr['Median Family Income'].astype(float).mean()
light_house_size = lr['Average Household Size'].astype(float).mean()
```

```python
# average median income and average household size in blocks that get 
# 'Vacant/Abandoned Buildings Reported' vs 'Alley Light Out'
ih_df = pd.DataFrame({'Avg Median Family Income': [build_income, light_income],\
                    'Avg Household Size': [build_house_size, light_house_size]},\
                    index=['Vacant/Abandoned Buildings Reported', 'Alley Light Out'])
ih_df
```
